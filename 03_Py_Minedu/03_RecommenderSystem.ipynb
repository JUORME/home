## Recomendadores Simples
Los recomendadores simples son sustemas básicos que recomiendan los mejores elementos en función de una determinada métrica o puntuación. 

```python
df.columns
```

```python
#Media del voto promedio de una escala de 0 a 10
media_puntaje = df['vote_average'].mean()
print(media_puntaje)
```

```python
#Calcular el minimo numero de votos 
conteo_votos = df['vote_count'].quantile(0.90)
print(conteo_votos)
```

```python
df_movies = df.copy().loc[df['vote_count'] >= conteo_votos]
print(df_movies.shape)
print(df.shape)
```

```python
# Funcion que calcula la calificacion ponderada de cada pelicula.
def calificacion(x, m= conteo_votos, c=media_puntaje):
    v = x['vote_count']
    r = x['vote_average']
    return (v/(v+m)*r) + (m/(m+v) * c)
```

```python
df_movies['puntaje'] = df_movies.apply(calificacion, axis= 1)
df_movies.head(3)
```

```python
#Ordenar los datos basado en el puntaje obtenido
df_movies = df_movies.sort_values('puntaje', ascending = False)
```
